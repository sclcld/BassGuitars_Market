```python
fender_clean= pd.DataFrame(
                            fender[[
                                    "Product Name", "Price", "Body Material","Neck Material", "Fingerboard Material", 
                                    "Number of Frets", "Nut Material", "Bridge Pickup", "Neck Pickup", "Bridge"
                                  ]])
fender_clean.columns= ["Product Name", "Price", "Body","Neck", "Fingerboard", "Frets", "Nut", "Bridge Pickup", "Neck Pickup", "Bridge"]
fender_clean["Manufacturer"] = ["Fender" for _ in range(len(fender_clean))]
fender_clean["Price"] = [float(price[:-5].replace(",",".")) for price in fender_clean["Price"]]
fender_clean["Fingerboard"]= [ENIT[fb.strip()] if not isinstance(fb, float)  else "nan" for fb in fender_clean["Fingerboard"]]
fender_clean["Nut"]= [ENIT[nut] if not isinstance(nut, float) else "nan" for nut in fender_clean["Nut"]]

squier_clean= pd.DataFrame(
                            squier[[
                                    "Product Name", "Price", "Body Material","Neck Material", "Fingerboard Material", 
                                    "Number of Frets", "Nut Material", "Bridge Pickup", "Neck Pickup", "Bridge"
                            ]])
squier_clean.columns= ["Product Name", "Price", "Body","Neck", "Fingerboard", "Frets", "Nut", "Bridge Pickup", "Neck Pickup", "Bridge"]
squier_clean["Manufacturer"] = ["Squier" for _ in range(len(squier_clean))]
squier_clean["Price"] = [float(price[:-5].replace(",",".")) for price in squier_clean["Price"]]
squier_clean["Neck"]= [ENIT[neck.strip()] if not isinstance(neck, float)  else "nan" for neck in squier_clean["Neck"]]
squier_clean["Fingerboard"]= [ENIT[fb.strip()] if not isinstance(fb, float)  else "nan" for fb in squier_clean["Fingerboard"]]
squier_clean["Nut"]= [ENIT[nut] if not isinstance(nut, float) else "nan" for nut in squier_clean["Nut"]]

hb_clean= pd.DataFrame(hb[["model", "price", "Body" , "Bolt-on neck", "Fingerboard", "Nut", "Pickup", "Bridge"]])
hb_clean.columns= ["Product Name", "Price", "Body","Neck", "Fingerboard", "Nut", "Neck Pickup", "Bridge"]
hb_clean["Manufacturer"] = ["Harley Benton" for _ in range(len(hb_clean))]
hb_clean["Price"]= [float(num) for num in hb_clean["Price"]]
hb_clean["Body"]= [ENIT[body.strip()] if not isinstance(body, float) else "nan" for body in hb_clean["Body"]]
hb_clean["Neck"]= [ENIT[neck.strip()] if not isinstance(neck, float)  else "nan" for neck in hb_clean["Neck"]]
hb_clean["Fingerboard"]= [ENIT[fb.strip()] if not isinstance(fb, float)  else "nan" for fb in hb_clean["Fingerboard"]]
hb_clean["Nut"]= [ENIT[nut.strip()] if not isinstance(nut, float) else "nan" for nut in hb_clean["Nut"]]

schechter_clean= pd.DataFrame(
                                schechter[[
                                            "Product Name","Price", "Body Material", "Neck Material", "Fretboard Material",
                                            "Frets","Nut", "Bridge Pickup", "Neck Pickup", "Bridge"
                                ]])

schechter_clean.columns= ["Product Name", "Price", "Body","Neck", "Fingerboard", "Frets", "Nut", "Bridge Pickup", "Neck Pickup", "Bridge"]
schechter_clean["Manufacturer"] = ["Schechter" for _ in range(len(schechter_clean))]
schechter_clean["Price"]= [float(p.replace("$","").replace(".00 to 59","")[:-4].replace(",","")) * 0.93 for p in schechter_clean["Price"]]
schechter_clean["Body"]= [ENIT[body.strip()] if not isinstance(body, float) else "nan" for body in schechter_clean["Body"]]
schechter_clean["Neck"]= [ENIT[neck.strip()] if not isinstance(neck, float)  else "nan" for neck in schechter_clean["Neck"]]
schechter_clean["Fingerboard"]= [ENIT[fb.strip()] if not isinstance(fb, float)  else "nan" for fb in schechter_clean["Fingerboard"]]
schechter_clean["Nut"]= [ENIT[nut.strip()] if not isinstance(nut, float) else "nan" for nut in schechter_clean["Nut"]]
schechter_clean["Frets"]= [fret.split()[0] if len(fret.split()) else fret > 1 for fret in schechter_clean["Frets"]]

fodera_clean= pd.DataFrame(fodera.drop(columns=["Unnamed: 0", "Top"]))
fodera_clean.columns= ["Product Name", "Price", "Body", "Neck", "Bridge Pickup", "Frets","Fingerboard"]
fodera_clean["Manufacturer"] = ["Fodera" for _ in range(len(fodera_clean))]
fodera_clean["Body"]= [ENIT[body.strip()] if not isinstance(body, float) else "nan" for body in fodera_clean["Body"]]
fodera_clean["Price"]= [float((price[:-3].replace(",",""))) * 0.93 for price in fodera_clean["Price"]]
fodera_clean["Neck"]= [ENIT[neck.strip()] if not isinstance(neck, float)  else "nan" for neck in fodera_clean["Neck"]]
fodera_clean["Fingerboard"]= [ENIT[fb.strip()] if not isinstance(fb, float)  else "nan" for fb in fodera_clean["Fingerboard"]]
```
